# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', 55)
```

# Loading the data

```python
df = pd.read_csv('UK-HPI-full-file-2021-11.csv')
df
```

# Filtering (period after 2005)

```python
df['Date'] = pd.to_datetime(df['Date'])  
df=df[df['Date']>='01/01/2005']
df
```

# House type analysis

```python
df.columns
```

```python
col = ['Date','RegionName', 'AreaCode', 'AveragePrice',
       'SalesVolume', 'DetachedPrice','SemiDetachedPrice','TerracedPrice', 'FlatPrice']
df=df[col]
df
```

```python
df = df[df['DetachedPrice'].notna()]
gb = df.groupby('Date').mean()
gb = gb.reset_index()
plt.plot(gb['Date'], gb['DetachedPrice'], label = 'Detached House price')
plt.plot(gb['Date'], gb['SemiDetachedPrice'],label = 'Semi-Detached House price')
plt.plot(gb['Date'], gb['TerracedPrice'],label = 'Terraced House price')
plt.plot(gb['Date'], gb['FlatPrice'],label = 'Flat price')
plt.legend()

```
